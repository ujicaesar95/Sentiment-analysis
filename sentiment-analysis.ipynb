# Define parallelize computing function

```python
# for multiprocessing support
from multiprocessing import  Pool, cpu_count
from functools import partial
from tqdm import tqdm
tqdm.pandas()

import numba
from os import cpu_count
n_cpu = cpu_count()
    
def parallelize(data, func, num_of_processes=n_cpu):
    data_split = np.array_split(data, num_of_processes)
    pool = Pool(num_of_processes)
    data = pd.concat(pool.map(func, data_split))
    pool.close()
    pool.join()
    return data
 
def run_on_subset(func, data_subset):
    return data_subset.apply(func)
 
def parallelize_on_rows(data, func, num_of_processes=n_cpu):
    return parallelize(data, partial(run_on_subset, func), num_of_processes)
```

# Declaration and variable initialization

```python
MAX_WORDS = 660
lr = 0.0001
epochs = 600
batch_size = 300
save_best=True
logger = logging.getLogger()

dirpath = os.getcwd()
dirname = Path(dirpath)
name = 'sentiment-analysis'
model_dir = 'models/'

stopwords = joblib.load(os.path.join(model_dir, 'stopwords.pkl'))
```

# Data preparation

```python
training_data = pd.read_csv('data/sentiment_analysis_training.csv')
test_data = pd.read_csv('data/sentiment_analysis_test.csv')
```

### Data cleaning

```python
def remove_stopwords(sentence):

    cleaned_words = [word for word in sentence.split() if word not in stopwords]

    return ' '.join(cleaned_words)
```

```python
training_data['REVIEW'] = training_data['REVIEW'].progress_apply(remove_stopwords)
test_data['REVIEW'] = test_data['REVIEW'].progress_apply(remove_stopwords)
```

### Data splitting

```python
X_train = training_data.loc[training_data['TAG'] == 'TRAIN']['REVIEW'].values
y_train = training_data.loc[training_data['TAG'] == 'TRAIN']['LABEL'].values
X_val = training_data.loc[training_data['TAG'].isin(['VALIDATE','TEST'])]['REVIEW'].values
y_val = training_data.loc[training_data['TAG'].isin(['VALIDATE','TEST'])]['LABEL'].values
X_test = test_data['REVIEW'].values
y_test = test_data['LABEL'].values
```

### Generating tokenizer

```python
tokenizer = Tokenizer(num_words=MAX_WORDS)
tokenizer.fit_on_texts(X_train)

joblib.dump(tokenizer, os.path.join(model_dir, '{}-tokenizer.pkl'.format(name)))
```

### Generating dictionary

```python
dictionary = tokenizer.word_index

with open(os.path.join(model_dir, '{}-dictionary.json'.format(name)), 'w') as dictionary_file:
    json.dump(dictionary, dictionary_file)
```

### Tokenizing training, validation and test data set

```python
def __convert_text_to_index_array(text):
        return [dictionary[word] for word in kpt.text_to_word_sequence(text) if word in dictionary.keys()]
```

```python
trainWordIndices = []
for text in X_train:
    wordIndices = __convert_text_to_index_array(text)
    trainWordIndices.append(wordIndices)
```

```python
valWordIndices = []
for text in X_val:
    wordIndices = __convert_text_to_index_array(text)
    valWordIndices.append(wordIndices)
```

```python
testWordIndices = []
for text in X_test:
    wordIndices = __convert_text_to_index_array(text)
    testWordIndices.append(wordIndices)
```

### Create one-hot matrices

```python
trainWordIndices = np.asarray(trainWordIndices)
valWordIndices = np.asarray(valWordIndices)
testWordIndices = np.asarray(testWordIndices)
```

### Create sequences to matrix

```python
X_train = tokenizer.sequences_to_matrix(trainWordIndices, mode='binary')
X_val = tokenizer.sequences_to_matrix(valWordIndices, mode='binary')
X_test = tokenizer.sequences_to_matrix(testWordIndices, mode='binary')
```

# Build model

```python
model = Sequential()
model.add(Dense(512, input_shape=(MAX_WORDS,), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(256, activation='sigmoid'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer=Adam(lr=lr), metrics=['accuracy'])
checkpointer = ModelCheckpoint(os.path.join(model_dir, '{}-model.hdf5'.format(name)),save_best_only=save_best, verbose=1)
logger.info('Success save model')
```

# Build metrics for visualize the result

```python
class Metrics(Callback):
            
    def on_train_begin(self, logs={}):
        self.val_precision = []
        self.val_recall = []
        self.val_f1score = []

    def on_epoch_end(self, batch, logs={}):
        X_val, y_val = self.validation_data[0], self.validation_data[1]
        y_pred = np.asarray(self.model.predict(X_val))
        y_val = np.argmax(y_val, axis=1)
        y_pred = np.argmax(y_pred, axis=1)
        self.val_precision.append(precision_score(y_val, y_pred, average='weighted'))   
        self.val_recall.append(recall_score(y_val, y_pred, average='weighted'))
        self.val_f1score.append(f1_score(y_val, y_pred, average='weighted'))
        print ("— val_precision: %f — val_recall %f — val_f1: %f " %(precision_score(y_val, y_pred, average='weighted'), recall_score(y_val, y_pred, average='weighted'), f1_score(y_val, y_pred, average='weighted')))
        return

    def get_precision(self):
        return self.val_precision

    def get_recall(self):
        return self.val_recall

    def get_f1score(self):
        return self.val_f1score
             
f1_metrics = Metrics()
```

# Training the model

```python
model_history = model.fit(X_train,
                            y_train,
                            batch_size=batch_size,
                            epochs=epochs,
                            validation_data=(X_val, y_val),
                            callbacks = [checkpointer, f1_metrics],
                            shuffle=True,
                            verbose=1)

model_structure = model.to_json()
with open(os.path.join(model_dir, '{}-model.json'.format(name)), 'w') as json_file:
    json_file.write(model_structure)

model_history.history['val_precision'] = f1_metrics.get_precision()
model_history.history['val_recall'] = f1_metrics.get_recall()
model_history.history['val_f1score'] = f1_metrics.get_f1score()

print('Finished Training')
```

# Visualize the result

```python
def evaluate_model(X_test, y_test, get_report=False, plot_res = False):
    COLOR = 'blue'
    mpl.rcParams['text.color'] = COLOR
    mpl.rcParams['axes.labelcolor'] = COLOR
    mpl.rcParams['xtick.color'] = COLOR
    mpl.rcParams['ytick.color'] = COLOR
    params = {'legend.fontsize': 20,
          'legend.handlelength': 2}
    plot.rcParams.update(params)
    
    _,accuracy=model.evaluate(X_test, y_test, verbose=1)
    if not get_report:
        print('Accuracy : {}'.format(accuracy))
    else:
        # Open the file
        with open('report.txt','w') as fh:
        # Pass the file handle in as a lambda function to make it callable
            model.summary(print_fn=lambda x: fh.write(x + '\n'))
            fh.write('\n')
            fh.write('\n')
            fh.write('Accuracy : {}'.format(accuracy)+ '\n')
    
    if plot_res:
        plt.figure(figsize=(20,10))
        plt.plot(model_history.history['acc'])
        plt.plot(model_history.history['val_acc'])
        plt.title('model accuracy, peak at {0:.2f}% of validation accuracy'.format(max(model_history.history['val_acc'])*100),
                  fontsize=30)
        plt.ylabel('accuracy',fontsize=28)
        plt.xlabel('epoch',fontsize=28)
        plt.legend(['train', 'test'], loc='upper left')
        plt.show()

        # summarize history for loss
        plt.figure(figsize=(20,10))
        plt.plot(model_history.history['loss'])
        plt.plot(model_history.history['val_loss'])
        plt.title('model loss, best at {0:.4f} of validation loss'.format(min(model_history.history['val_loss'])),fontsize=30)
        plt.ylabel('loss',fontsize=28)
        plt.xlabel('epoch',fontsize=28)
        plt.legend(['train', 'test'], loc='upper left')
        plt.show()
```
